# A Summary of Resources for Query Letters

## A Pathway to Success

The General Consensus from resources I've reviewed is that you will be writing they query lette for a long time, and therefore it is prudent to break your efforts into 3-5 five areas that you work on once a week:

1. Agent Research - You need to know what agents you are looking for. Find agents by one of two ways
    - Read books to figure out your favorite author's agents
    - Go online to look for agents in your *genre categories*
    - Agents should be *accepting submissions* when you contact them
    
2. Query Letter Development - In order to send query letters, you need to have one. One you have looked at a suitable list of agents, you should spend time personalizing them 
    - In query tracker, there are some other fields you would submit multiple times (like books, etc). You should build out your list default answers to commonly asked questions to agents.

3. Writing Sample Development - After the agent has taken interest in your work, you have to send them a writing sample. You don't want the writing sample to be a bummer compared to your query letter. 

4. Continued Pursuit of Writing - Sometimes, the first book isn't always the one that gets published. It's good to engage your creativity by working on multiple works.

5. Continued persuit of Reading - staying up to date on popular books in your *genre categories*
    - Try going to book stores and asking them owner about their popular genre categories. 
    - Look at publishing houses you like and see what they've published this year.

### Frequently Asked Questions (AKA Questions I asked other people)
**Q:** How many agents will I query

**A:** 30-300

**Q:** Do agents talk to each other? How much can I reuse queries

**A:** Agents do talk to each other. Do no reuse queries if possible, especially names, or confuse books represented by different agencies

### Elements of a query letter

|3 Element Approach| 4 Element Approach | 6 Element Approach |
|----------|----------|-------|
| Hook | Agent Appeal | Hook |
| Pitch| Hook| Pitch |
| Bio| Comps| Agent Appeal|
|  -  | Bio| Comps | Comps |
|   -  |      -  | Bio |
|    -  |  -    | Marketing Plan |

There are several ways to look at the query letter. Given above are certain approaches that I've come across for writing query letters.

* *Hook* - The hook seems to be the words or phrase that gets people interested in your book. It's very sort and sets the stage for the rest of the letter. In some versions of the query letter architecture, it includes Agent Appeal, or Comps. 
* *Pitch* - This is a longer section that tells more about your book. People have different opinions about this length. Some like it shorter, and less like a synopsis. 
* *Agent Appeal* - This is a personalized section to the agent
* *Comps* - Comparable titles
* *Bio* - This should be longer than you think, probably. 
* *Marketing plan* - Your audience, followers on instagram, etc. Seems to be more popular with non-fiction.

### Frequently Asked Questions (AKA Questions I asked other people)
**Q:** How long should the pitch be?

**A:** I've recieved varying feedback on this. Some authors I've ask say that the pitch should go *up to the middle of your novel*. Others prefer that to be in the synopsis. It is worth nothing that fiction books typically use the synopsis less when querying.

**Q:** Should the pitch be about the plot or essence of the novel

**A:** A plot would be "Person A discovers a new quantum singularity and finds an alternative dimision", while an essence is like "This is an exploration of disillusionment with marriage based around the viewing of another version of youself's life". It seems like both are important, but the plot is more important for fiction query letters.

## 3. Writing Sample Development 

In fiction land the agent will typically ask for the initial sections of your book. It's good to get feedback on your work, and it can be daunting to have people read your work; however, the best people to help you are your friends and community groups! 

### Writing for friends
The best harsh feedback I've gotten from my friends, I've put in bullets

* Don't give me unfinished work: 
    * Try not give people a draft you haven't revised at least a couple times. Your friends don't want to spend time tracking down spelling errors and finding unfinished sentences, they want the good stuff. Revise twice before you enlist someone in your life to help you
* I don't know how to edit, so I don't want to do it
    * Most people haven't taken a lit class, so they don't know what you're looking for. Editing is a chore for your friends, and they're doing it because they love you. Give them a hand and tell them exactly what you are looking for. Here are some suggestions that I've started distrubting with my work when I ask people to edit, and it helps them a lot.
        * *Clarity: A good comment is clear about the nature of their comment, trying to be as specific as possible*
            - *Example of an unclear comment: “I would rework the three paragraphs before this section”*
            - *Better comment: I’m not sure who is talking during the above scene. I think the person talking is X or Y, but not sure*
        * *Contains a resolution: A great comment indicates a proposal to fix the problem*
            - *Example of a comment without a solution: I’m not sure who is talking during the above scene. I think the person talking is X or Y, but not sure*
            - *Example of a comment with a proposed solution: I’m not sure who is talking during the above scene. If you could add a character name to the first and third line of dialogue that would be really helpful*
        * *Location: A stellar comment occurs During or Before the location of impact. This is because I typically edit the novel in sequence order, so if a comment takes place AFTER it's reference point, I have to go back and re-edit a previously completed section*
            - *Example of a comment occurring after the fact: I’m not sure who is talking during the above scene. If you could add a character name to the first and third line of dialogue that would be really helpful*
            - *Supreme, gold-star, ultra-useful comment: In the next three paragraphs, I’m not sure who is talking. If you could add a character name to the first and third line of dialogue that would be really helpful*

Some comments I've heard of why writers don't ask their friends for help
* My friends don't have good taste in literature, and don't know how to write
    * In my field, there is something called an SME (Subject Matter Expert). The subject matter expert knows everything about their field. Every single one of your friends is an SME in something, even if they are not an expert in literature. I've had friends who are in the government give me feedback on my description of classified information. I've had biologists tell me what I've gotten wrong about their field. The more people you have read the book, the more SMEs you will engage with. It makes your book seems smarter and more educated.
* I'm worried my friends will make fun of my writing.
    * Yeah, they might, but you need to know that. If your writing is bad, the agent won't pick it up, so you should know before hand.

## 5. Continued persuit of Reading 

Reading is how you learn about agents and what is popular in your field. You want to understand what is going in your genre or subgenre. Is Grimdark still a work people use? If you're writing Grimdark, who is publishing it, and who are the agents. Can you add something to your list of like books. 

### Doing the research

* Going to local bookstores
    * There are some really great book stores that have curated lists. One that comes to mind is Powell's books in Oregon, but the experience is so much better in person. If you are far away and out of state, maybe you can have a friend go there and find your genres. You want to pay attention to how the book store describes your book, and use that language to help sell it.
    * Talk to the book store's owner about what is popular
* Going to publisher's pages
    * Self-explanatory
* Go to the library
    * The library isn't so much concerned with the supply and demand of books or selling titles, but they do know what people are enjoying right now. You can always check out the new book section, or talk to a librarian about what is hot in your genre.
